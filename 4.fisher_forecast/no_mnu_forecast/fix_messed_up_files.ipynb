```python
import os
import datetime


# Specify the directory path and call the function
directory_path = "/oak/stanford/orgs/kipac/users/delon/aemulusnu_massfunction/"

# Initialize an empty list to store the file names
recent_files = []

# Get the current time
current_time = datetime.datetime.now()

# Get the list of all files in the directory
files = os.listdir(directory_path)

# Loop through the files
for file in files:
    file_path = os.path.join(directory_path, file)

    # Ensure the path is a file
    if os.path.isfile(file_path):
        # Get the file modification time
        file_mod_time = datetime.datetime.fromtimestamp(os.path.getmtime(file_path))

        # Calculate the time difference
        time_diff = current_time - file_mod_time
            

        # If the time difference is less than 3 hours, print the file name
        if (time_diff.total_seconds() < 8 * 60 * 60 and 'b16' not in file and '_st' not in file):
            recent_files.append(file)
```

```python
for fname in sorted(recent_files):
    print(directory_path+fname)
    print('\t',directory_path+fname+'_no_mnu')
```

```python
def rename_file(old_name, new_name):
    """
    Rename a file from 'old_name' to 'new_name'.
    """
    try:
        os.rename(old_name, new_name)
        print(f"File renamed from {old_name} to {new_name}")
    except FileNotFoundError:
        print(f"The file {old_name} was not found.")
    except PermissionError:
        print("Permission denied. Cannot rename the file.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
for fname in sorted(recent_files):
    old_name = directory_path + fname
    new_name = directory_path + fname+'_no_mnu'
    rename_file(old_name, new_name)
```
