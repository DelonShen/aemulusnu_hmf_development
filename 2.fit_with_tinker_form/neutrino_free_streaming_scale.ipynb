```python
import numpy as np
import pyccl as ccl

def neutrino_free_streaming_scale(cosmo, a):
    #1212.6154 Eq. 33
    z = scaleToRedshift(a)
    
    
    mnu = cosmo['m_nu'][0] #ev
    Omega_m = cosmo.omega_x(1.0, 'matter')
    Omega_L = cosmo.omega_x(1.0, 'dark_energy')
    
    return 8*(1+z)/np.sqrt(Omega_L + Omega_m*(1+z)**3) * 1 / mnu #h^-1 Mpc

import pickle
cosmos_f = open('../data/cosmo_params.pkl', 'rb')
cosmo_params = pickle.load(cosmos_f) #cosmo_params is a dict
cosmos_f.close()

from aemulusnu_massfunction.emulator_training import *

for box in cosmo_params:
    print(box)
    cosmo = (get_ccl_cosmology(tuple(get_cosmo_vals(cosmo_params[box]))))
    
    
    # Calculate the free streaming scale at different scale factors
    a_values = np.linspace(0.3, 1, 5)
    fs_scales = neutrino_free_streaming_scale(cosmo, a_values)

    # Print results
    for a, fs_scale in zip(a_values, fs_scales):
        print(f"Scale factor: {a:.3f}, Free streaming scale: {fs_scale:.2f} Mpc/h")
    print()
```
